```python
#Importez la librairie Pandas
import pandas as pd
```

```python
#Chargez le fichier "Produits.xlsx" de données dans un objet appelé "data"
data = pd.read_excel('Produits.xlsx')
data
```

```python
#Copiez data dans un autre object appelé df
df = data.copy()
```

```python
#Affichez les informations concernant l'objet df
df
```

```python
#En vous basant sur les résultats de la question précédente, Combien y a-t-il d’observations et de variables dans le fichier 
#à travers les questions précédentes , nous constactons qu'il ya 77 lignes et 7 colonnes qui répresentent les observations .et au niveau des variables nous avons:Nom;Categorie;Origine;Ville;Prix;Stock;Valeur
df.shape
df.columns
```

```python
#Affichez les 10 premières lignes des données de l'objet df
df.head(10)
```

```python
#Supprimer la colonne "Valeur"
df=df.drop(['Valeur'], axis=1)
df
```

```python
#Afficher les statistiques descriptives pour les variables quantitatives de l'objet df
df.describe()
```

```python
#Afficher les statistiques descriptives pour les variables qualitatives de l'objet df
df.describe(include='object')
```

```python
#On veut afficher les produits dont la catégorie = boissons
df[(df['Categorie'] == 'Boissons')]
```

```python
#On veut afficher les produits dont la Catégorie = boissons et le prix >100
produits=df[(df['Categorie']=='Boissons') & (df['Prix']>100)]
produits
```

```python
#On veut afficher les produits dont la Catégorie = boissons et l'origine=CEE et prix > 100
produits1=df[(df['Categorie']=='Boissons') & (df['Origine']=='CEE') & (df['Prix']>100)]
produits1
```

```python
#On veut afficher les produits dont la (Catégorie = boissons et origine = CEE) OU (catégorie = condiment)
produits2=df[((df['Categorie']=='Boissons') & (df['Origine']=='CEE'))|(df['Categorie']=='Condiments')]
produits2
```

```python
#On veut afficher les produits dont (Catégorie = viande ET origine = CEE) OU (catégorie = condiment ET origine = extérieur)
produits3=df[((df['Categorie']=='Viandes') & (df['Origine']=='CEE')) | ((df['Categorie']=='Condiments') & (df['Origine']=='Exterieur'))]
produits3
```

```python
#On veut afficher les produits dont le prix > 70 et prix <=100
produits4 = df[(df['Prix']>70) & (df['Prix']<=100)]
produits4
```

```python
#Lister les aliments dont le prix est compris entre 100 et 200, et qui sont des « viandes »
liste_aliments=df[((df['Prix']>=100) & (df['Prix']<=200) & (df['Categorie']=='Viandes'))]
liste_aliments
```

```python
#Lister les 15 produits les moins chers
produits_moins_chers=df.sort_values(by='Prix').head(15)
produits_moins_chers
```

```python
#Calculer la moyenne de prix des boissons distribuées à Lyon (pivot_table)
moyenne_prix=pd.pivot_table(df, values ='Prix', index=['Ville'],columns=['Categorie']).loc['Lyon','Boissons']
moyenne_prix
```

```python
#Quels sont les 5 produits les moins chers vendus à Lyon ?
pl=df.sort_values(by='Prix')
pl[pl['Ville']=='Lyon'].head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
# afficher les variables quantitatives
df.select_dtypes(include=['int','float']).columns
list_quan=['Prix', 'Stock']
list_quan
```

```python
plt.figure(figsize=(15,7))
i=1
for element in list_quan:
    plt.subplot(1,2,i)
    df[element].value_counts().plot(kind='hist')
    i+=1
    plt.title(element)
```
